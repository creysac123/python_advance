### Q3. What is a histogram, exactly? Name a numpy method for creating such a graph.



A histogram is a graphical representation of the distribution of a dataset. It consists of a series of bins that represent different ranges of values, and the height of each bin corresponds to the frequency or count of values falling within that range. It provides insights into the underlying data distribution, such as the central tendency, spread, and shape.

In NumPy, the method numpy.histogram() can be used to create a histogram. It takes an array of data and returns the counts and bin edges for each bin in the histogram. The resulting histogram can then be visualized using plotting libraries like Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate some data for this demonstration.
data = np.random.randn(1000)

# Fit a normal distribution to the data:
mu, std = norm.fit(data)

# Plot the histogram.
plt.hist(data, bins=25, density=True, alpha=0.6, color='g',edgecolor='black')

# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)

plt.show()
```

### Q4. If necessary, how do you change the aspect ratios between the X and Y axes?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate some data for this demonstration.
data = np.random.randn(1000)

# Fit a normal distribution to the data:
mu, std = norm.fit(data)

# Create a new figure and add a new axes to it.
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# Plot the histogram.
ax.hist(data, bins=25, density=True, alpha=0.6, color='g')

# Plot the PDF.
xmin, xmax = ax.get_xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
ax.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
ax.set_title(title)

# Change the aspect ratio of the axes.
ax.set_aspect('equal')

plt.show()
```

### Q5. Compare and contrast the three types of array multiplication between two numpy arrays: dot product, outer product, and regular multiplication of two numpy arrays.

The three types of array multiplication in NumPy are:

- Dot Product: Scalar value obtained by multiplying corresponding elements of two arrays and summing the results. It follows matrix multiplication rules and is useful for similarity measures and solving linear equations.

- Outer Product: Matrix obtained by multiplying each element of the first array with every element of the second array. It captures pairwise interactions between elements and is commonly used in computations involving tensors.

- Regular Multiplication: Element-wise multiplication of corresponding elements in two arrays. The resulting array has the same shape as the input arrays and is useful for element-wise operations like scaling and broadcasting.

### Q7. Can string data be stored in numpy arrays? If so, list at least one restriction that applies to this data.



String data can be stored in numpy arrays, but there are some restrictions. The main restrictions are that the strings must have fixed lengths, there is a maximum length limit, and all elements in the array must have the same data type. Additionally, string operations in numpy arrays can be slower and may require more memory compared to numerical computations. Despite these limitations, numpy arrays can still be useful for working with string data in combination with other numerical data.
